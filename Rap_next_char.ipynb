```python
import pandas as pd
import numpy as np
import random
```

```python
f = open('lyrics1.txt','r')
ds = f.read()
f.close()
```

```python
data = ' '.join([ix.strip() for ix in ds.split('\n')]).lower()
```

```python
f = open('lyrics2.txt','r')
ds = f.read()
f.close()
```

```python
data +=' '.join([ix.strip() for ix in ds.split('\n')]).lower()
```

```python
dictn = {}
ordr = 4

for ix in range(len(data)-ordr):
    
    ctx = data[ix:ix+ordr]
    
    next_char = data[ix+ordr]
    
    if dictn.get(ctx) is None:
        dictn[ctx]={}
        dictn[ctx][next_char]=1
    else:
        if dictn[ctx].get(next_char) is None:
            dictn[ctx][next_char] = 1
        else:
            dictn[ctx][next_char]+=1

```

```python
## Making the counts of chaarachter as the probability and normalising the counts between 0 to 1
```

```python
for ix in dictn.keys():
    
    cnt = float(sum(dictn[ix].values()))
    
    for kx in dictn[ix].keys():
        dictn[ix][kx] = dictn[ix][kx]/cnt
        
        
```

```python
## accurences are changed as pobabilities
dictn
```

```python
def next_generate(ctx):
    possible = dictn.get(ctx)
    
    if possible is None:
        return " "
    
    keys = possible.keys()
    random.shuffle(keys)
    probs = [possible[ix] for ix in keys]
    
    char = np.random.choice(keys,p=probs)
    return char
```

```python
word = 'hell'
sentence = ''+word

for ix in range(1000):
    new_char = next_generate(word)
    sentence = sentence + new_char
    word = sentence[-ordr:]
print(sentence)
```
